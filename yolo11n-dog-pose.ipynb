```python
from ultralytics import YOLO
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# í•™ìŠµ ì™„ë£Œëœ ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
model = YOLO("runs/pose/train/weights/best.pt")

# ì´ë¯¸ì§€ í´ë” ê²½ë¡œ (ì‹¤ì œ ì´ë¯¸ì§€ ê²½ë¡œì— ë§žê²Œ ìˆ˜ì •í•˜ì„¸ìš”)
image_folder = "images"
image_files = sorted([f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png'))])[:11]  # 11ìž¥ ì‚¬ìš©

# YOLO-Pose ì‹¤í–‰ ë° ê²°ê³¼ ì‹œê°í™”
for image_file in image_files:
    image_path = os.path.join(image_folder, image_file)
    image = cv2.imread(image_path)

    if image is None:
        print(f"ì´ë¯¸ì§€ë¥¼ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ì—†ìŒ: {image_path}")
        continue

    # ëª¨ë¸ ì˜ˆì¸¡ ìˆ˜í–‰ (ì´ë¯¸ì§€ ê²½ë¡œ ë˜ëŠ” ì´ë¯¸ì§€ë¥¼ ì§ì ‘ ì „ë‹¬ ê°€ëŠ¥)
    results = model(image_path)

    for result in results:
        # result.boxes.xyxy : bounding box ì¢Œí‘œ ë¦¬ìŠ¤íŠ¸
        # result.boxes.cls  : ê° bounding boxì˜ í´ëž˜ìŠ¤ ì¸ë±ìŠ¤ ë¦¬ìŠ¤íŠ¸
        # result.keypoints.xy : ê° ê°ì²´ì˜ keypoints ì¢Œí‘œ ë¦¬ìŠ¤íŠ¸

        # ì¸ë±ìŠ¤ë¥¼ ì´ìš©í•˜ì—¬ bounding box, keypoints, í´ëž˜ìŠ¤ ì •ë³´ë¥¼ í•¨ê»˜ ì²˜ë¦¬
        for i, (box, keypoints) in enumerate(zip(result.boxes.xyxy, result.keypoints.xy)):
            x1, y1, x2, y2 = map(int, box)

            # í´ëž˜ìŠ¤ ì¸ë±ìŠ¤ë¥¼ ì–»ê³ , model.names ë”•ì…”ë„ˆë¦¬ì—ì„œ í•´ë‹¹ í´ëž˜ìŠ¤ ì´ë¦„ì„ ê°€ì ¸ì˜´
            class_id = int(result.boxes.cls[i])
            label = model.names.get(class_id, str(class_id))

            # Bounding Box ê·¸ë¦¬ê¸°
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                        0.5, (255, 0, 0), 2)

            # Keypoint ê·¸ë¦¬ê¸°
            for x, y in keypoints:
                if x > 0 and y > 0:  # ìœ íš¨í•œ ì¢Œí‘œì¸ ê²½ìš°ì—ë§Œ ê·¸ë¦¼
                    cv2.circle(image, (int(x), int(y)), 5, (0, 255, 0), -1)

    # BGR -> RGB ë³€í™˜ í›„ matplotlibë¥¼ ì´ìš©í•˜ì—¬ ê²°ê³¼ ì¶œë ¥
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 10))
    plt.imshow(image_rgb)
    plt.title(f"YOLO-Pose Keypoints - {image_file}")
    plt.axis("off")
    plt.show()
```

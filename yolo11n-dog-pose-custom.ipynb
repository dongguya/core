#### 3.1 Ïù¥ÎØ∏ÏßÄ

```python
from ultralytics import YOLO
import cv2
import os
import random
import matplotlib.pyplot as plt

model = YOLO("runs/pose/dog-pose-custom2/weights/best.pt")

image_folder = "images"
all_image_files = [f for f in os.listdir(image_folder) if f.lower().endswith(('.jpg', '.png'))]

num_samples = 8
sample_files = random.sample(all_image_files, num_samples) if len(all_image_files) >= num_samples else all_image_files

fig, axes = plt.subplots(2, 4, figsize=(10, 5))
axes = axes.flatten()

for idx, image_file in enumerate(sample_files):
    image_path = os.path.join(image_folder, image_file)
    image = cv2.imread(image_path)
    if image is None:
        continue

    # verbose=FalseÎ°ú ÏÑ§Ï†ïÌï¥ Î°úÍ∑∏ ÏµúÏÜåÌôî
    results = model.predict(source=image_path, verbose=False)

    for result in results:
        for i, (box, keypoints) in enumerate(zip(result.boxes.xyxy, result.keypoints.xy)):
            x1, y1, x2, y2 = map(int, box)
            class_id = int(result.boxes.cls[i])
            label = model.names.get(class_id, str(class_id))

            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
            label_x = x1 + 5
            label_y = y1 + 20
            cv2.putText(image, label, (label_x, label_y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)

            for x, y in keypoints:
                if x > 0 and y > 0:
                    cv2.circle(image, (int(x), int(y)), 5, (0, 255, 0), -1)

    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axes[idx].imshow(image_rgb)
    axes[idx].set_title(f"YOLO11n-dog-pose-custom - {image_file}", fontsize=6)
    axes[idx].axis("off")

plt.tight_layout()
plt.show()
```

```python
import yt_dlp

channel_url = "https://www.youtube.com/shorts/YzW8PZ8PLpc"
output_path = "videos"
extracter = yt_dlp.YoutubeDL({"outtmpl": f"{output_path}/%(id)s.%(ext)s"})
extracter.download([channel_url])
```

```python
import cv2
from ultralytics import YOLO
import os

# 1. Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞
model = YOLO("runs/pose/dog-pose-custom2/weights/best.pt")

# 2. ÎèôÏòÅÏÉÅ ÌååÏùº Í≤ΩÎ°ú Î∞è Ï∂úÎ†• ÏÑ§Ï†ï
filename = "YzW8PZ8PLpc"
input_video_path = f"/content/videos/{filename}.webm"
output_video_path = f"/content/videos/{filename}.mp4"

cap = cv2.VideoCapture(input_video_path)

# ÎèôÏòÅÏÉÅ Ï†ïÎ≥¥ Ï∂îÏ∂ú (ÌîÑÎ†àÏûÑ ÎÑàÎπÑ, ÎÜíÏù¥, FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = cap.get(cv2.CAP_PROP_FPS)

# 3. VideoWriter ÏÑ§Ï†ï (MP4Î°ú Ï†ÄÏû•)
fourcc = cv2.VideoWriter_fourcc(*"mp4v")  # ÏΩîÎç± ÏÑ§Ï†ï
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        # Îçî Ïù¥ÏÉÅ ÌîÑÎ†àÏûÑÏù¥ ÏóÜÏúºÎ©¥ Ï§ëÎã®
        break

    # 4. Î™®Îç∏ ÏòàÏ∏° (ÌîÑÎ†àÏûÑ Îã®ÏúÑ)
    results = model.predict(source=frame, verbose=False)

    # 5. ÏòàÏ∏° Í≤∞Í≥ºÎ•º ÌîÑÎ†àÏûÑÏóê ÏãúÍ∞ÅÌôî
    for result in results:
        for i, (box, keypoints) in enumerate(zip(result.boxes.xyxy, result.keypoints.xy)):
            x1, y1, x2, y2 = map(int, box)

            # ÌÅ¥ÎûòÏä§ Ïù∏Îç±Ïä§ÏôÄ ÎùºÎ≤®
            class_id = int(result.boxes.cls[i])
            label = model.names.get(class_id, str(class_id))

            # Î∞îÏö¥Îî© Î∞ïÏä§
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)

            # ÎùºÎ≤® ÌÖçÏä§Ìä∏
            label_x = x1 + 5
            label_y = y1 + 20
            cv2.putText(frame, label, (label_x, label_y),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)

            # ÌÇ§Ìè¨Ïù∏Ìä∏
            for x, y in keypoints:
                if x > 0 and y > 0:  # Ïú†Ìö®Ìïú Ï¢åÌëúÎßå ÌëúÏãú
                    cv2.circle(frame, (int(x), int(y)), 5, (0, 255, 0), -1)

    # 6. ÏãúÍ∞ÅÌôîÍ∞Ä ÎÅùÎÇú ÌîÑÎ†àÏûÑÏùÑ VideoWriterÏóê Ï†ÄÏû•
    out.write(frame)
    frame_count += 1

cap.release()
out.release()

print(f"Ï≤òÎ¶¨Îêú Ï¥ù ÌîÑÎ†àÏûÑ Ïàò: {frame_count}")
print(f"Í≤∞Í≥º ÎèôÏòÅÏÉÅÏù¥ '{output_video_path}'Î°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")
```
